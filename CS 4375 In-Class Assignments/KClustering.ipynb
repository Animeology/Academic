### Clustering 
### Data Prepation
### Unsupervised learning reminder: no neet to train test split no label to test against

### we do need to transform categorical features into numeric ones where makes sense to do so , as well as 
### scaling the data due to distance being a key factor

```python
X = pd.get_dummies(df)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
scaled_X
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2)
```

```python
cluster_labels = model.fit_predict(scaled_X)
```

```python
cluster_labels
```

```python
len(scaled_X)
```

```python
len(cluster_labels)
```

```python
X['Cluster'] = cluster_labels
```

```python
sns.heatmap(X.corr())
```

```python
plt.figure(figsize=(12,6), dpi=200)
X.corr()['Cluster'].iloc[:-1].sort_values().plot(kind='bar')
```

```python
# al bow method
ssd = []
for k in range(2,10):
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,1), ssd, 'o--')
plt.xlabel("K value")
plot.ylabel("Sum of Squared Distance")
```
